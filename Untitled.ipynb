```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
cols=["fLength","fWidth","fSize","fConc","fConc1","fAsym","fM3Long","fM3Trans","fAlpha","fDist","class"]
df=pd.read_csv("magic04.data",names=cols)
df
```

```python
df["class"]=(df["class"]=="g").astype(int)
```

```python
for label in cols:
    plt.hist(df[df['class']==1][label].head(20), color='blue', label='gamma', alpha=0.7 ,density=True)
    plt.hist(df[df['class']==0][label].head(20), color='red', label='hadron', alpha=0.7 ,density=True)
    plt.title(label)
    plt.xlabel(label)
    plt.ylabel('Probability')
    plt.legend()
    plt.show()
```

```python
#splitting dataset into training, validation and test sets
train,validation,test=np.split(df.sample(frac=1),[int(0.6*len(df)),int(0.8*len(df))])
```

```python
print('train data:',len(train) ,' ', 'validation set:',' ',len(validation),'test set:',' ',len(test))
```

```python
#check correlation in number of counts in the training set
print("gamma particles:",len(train[train['class']==1]))
print("gamma particles:",len(train[train['class']==0]))
```

```python
def scale_dataset(dataframe,oversample=False):
    X=dataframe[dataframe.columns[:-1]].values
    y=dataframe[dataframe.columns[-1]].values

    scaler=StandardScaler()
    X=scaler.fit_transform(X) #scale input features

    if oversample:
        ros=RandomOverSampler()
        X,y=ros.fit_resample(X,y) # perform oversampling

    data=np.hstack((X,(np.reshape(y,(-1,1))))) #combine columns of array X and that of vector y to create a new array 'data'

    return data,X,y
```

```python
train,X_train,y_train=scale_dataset(train,oversample=True) #perform oversampling in training set only
validation,X_validation,y_validation=scale_dataset(validation,oversample=False)
test,X_test,y_test=scale_dataset(test,oversample=False)
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

```python
knn_predictions=knn_model.predict(X_test)
knn_predictions
```

```python
y_test
```

```python
print(classification_report(y_test,knn_predictions))
```
