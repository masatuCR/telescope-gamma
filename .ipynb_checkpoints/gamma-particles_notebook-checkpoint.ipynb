```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
cols=["fLength","fWidth","fSize","fConc","fConc1","fAsym","fM3Long","fM3Trans","fAlpha","fDist","class"]
df=pd.read_csv("magic04.data",names=cols)
df
```

```python
df["class"]=(df["class"]=="g").astype(int)
```

```python
df.tail(20)
df['class'].value_counts()
```

```python
for label in cols:
    plt.hist(df[df['class']==1][label].head(20), color='blue', label='gamma', alpha=0.7 ,density=True)
    plt.hist(df[df['class']==0][label].head(20), color='red', label='hadron', alpha=0.7 ,density=True)
    plt.title(label)
    plt.xlabel(label)
    plt.ylabel('Probability')
    plt.legend()
    plt.show()
```

```python
#splitting dataset into training, validation and test sets
train,validation,test=np.split(df.sample(frac=1),[int(0.6*len(df)),int(0.8*len(df))])
```

```python
print('train data:',len(train) ,' ', 'validation set:',' ',len(validation),'test set:',' ',len(test))
```

```python
#check correlation in number of counts in the training set
print("gamma particles:",len(train[train['class']==1]))
print("gamma particles:",len(train[train['class']==0]))
```

```python
def scale_dataset(dataframe,oversample=False):
    X=dataframe[dataframe.columns[:-1]].values
    y=dataframe[dataframe.columns[-1]].values

    scaler=StandardScaler()
    X=scaler.fit_transform(X) #scale input features

    if oversample:
        ros=RandomOverSampler()
        X,y=ros.fit_resample(X,y) # perform oversampling

    data=np.hstack((X,(np.reshape(y,(-1,1))))) #combine columns of array X and that of vector y to create a new array 'data'

    return data,X,y
```

```python
train,X_train,y_train=scale_dataset(train,oversample=True) #perform oversampling in training set only
validation,X_validation,y_validation=scale_dataset(validation,oversample=False)
test,X_test,y_test=scale_dataset(test,oversample=False)
```

```python
#k-nearest neighbor
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
knn_predictions=knn_model.predict(X_test)
print(classification_report(y_test,knn_predictions))
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
naive_model=GaussianNB()
naive_model.fit(X_train,y_train)#train naiveBayes model
naive_predictions=naive_model.predict(X_test) #make predictions using Naive_Bayes algorithm
print(classification_report(y_test,naive_predictions))#create classification report
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_model=LogisticRegression()    #define logistic regression
logistic_model.fit(X_train,y_train)    #train the logistic regression model
logistic_predictions=logistic_model.predict(X_test)    #make predictions
print(classification_report(y_test,logistic_predictions))  #generate classification report
```

Support Vector Machines

```python
#Support Vector Machines
from sklearn.svm import SVC
supportVector_model=SVC()
supportVector_model.fit(X_train,y_train)
supportVector_predictions=supportVector_model.predict(X_test)
print(classification_report(y_test,supportVector_predictions))
```

```python
#Neural Network(using TensorFlow Library)
```

```python
#draw the two graphs separately
def plot_loss(history):
    plt.plot(history.history['loss'],label='loss')
    plt.plot(history.history['val_loss'],label='val_loss')
    plt.xlabel('Epochs')
    plt.ylabel('binary crossentropy')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_accuracy(history):
    plt.plot(history.history['accuracy'],label='accuracy')
    plt.plot(history.history['val_accuracy'],label='val_accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import tensorflow as tf
neural_network_model=tf.keras.Sequential([
    tf.keras.layers.Dense(32,activation='relu',input_shape=(10,),),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
neural_network_model.compile(optimizer=tf.keras.optimizers.Adam(0.001),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=neural_network_model.fit(X_train,y_train,batch_size=32,epochs=100,validation_split=0.2)
```

```python
plot_loss(history)
plot_accuracy(history)
```

```python
#Draw the graphs side by side as two subplots(Alternative approach of drawing the same two graphs drawn above)
def plot_loss_and_accuracy(history):
    fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
    ax1.plot(history.history['loss'],label='loss')
    ax1.plot(history.history['val_loss'],label='val_loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('binary crossentropy')
    ax1.legend()
    ax1.grid(True)
    ax2.plot(history.history['accuracy'],label='accuracy')
    ax2.plot(history.history['val_accuracy'],label='val_accuracy')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('accuracy')
    ax2.legend()
    ax2.grid(True)
    plt.show()

plot_loss_and_accuracy(history)
```

```python
#function to train a model that accepts parameters number of nodes and learning rate etc
def train_model(X_train,y_train,num_nodes,dropout_prob,learning_rate,batch_size,epochs):
    neural_network_model=tf.keras.Sequential([
    tf.keras.layers.Dense(num_nodes,activation='relu',input_shape=(10,),),
    tf.keras.layers.Dropout(dropout_prob),
    tf.keras.layers.Dense(num_nodes,activation='relu'),
    tf.keras.layers.Dropout(dropout_prob),
    tf.keras.layers.Dense(1,activation='sigmoid')
    ])
    neural_network_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate),loss='binary_crossentropy',metrics=['accuracy'])
    history=neural_network_model.fit(X_train,y_train,batch_size=batch_size,epochs=epochs,validation_split=0.2)
    return neural_network_model,history
```

```python
#now lets try different parameters (learning rate, dropout probability and batch size etc) passed in a nested function
#mind you this will take time to complete 
epochs=100
least_validation_loss=float('inf')
least_loss_neural_network_model=None
for num_nodes in [16,32,64]:
    for learning_rate in [0.01,0.05]:
        for dropout_prob in [0,0.2]:
            for batch_size in [32,16,24]:
                print(f"number of nodes:{num_nodes}, learning rate:{learning_rate},dropout probability:{dropout_prob} and batch size:{batch_size}")
                neural_network_model,history=train_model(X_train,y_train,num_nodes,dropout_prob,learning_rate,batch_size,epochs)
                plot_loss_and_accuracy(history)
                validation_loss=neural_network_model.evaluate(X_validation,y_validation)[0]
                if validation_loss < least_validation_loss:
                    least_validation_loss = validation_loss
                    least_loss_neural_network_model = neural_network_model
```
